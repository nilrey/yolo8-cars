```python
from ultralytics import YOLO
```

```python
import cv2
```

```python
model = YOLO('yolov8n.pt')
```

```python
# Вывод всех классов
print("Классы, определяемые моделью:")
for class_id, class_name in model.names.items():
    print(f"{class_id}: {class_name}")
```

```python
video_path = '/content/drive/MyDrive/Colab Notebooks/cars/input/s-001.mp4'
# Обработка видео
cap = cv2.VideoCapture(video_path)
output_path = '/content/drive/MyDrive/Colab Notebooks/cars/output/cars-video-001-s-out-01.mp4'
```

```python
fps = int(cap.get(cv2.CAP_PROP_FPS))
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

allowed_indices = {2, 3, 5, 6, 7, 8}
```

```python
print(fps)
```

```python
out = cv2.VideoWriter(output_path, cv2.VideoWriter_fourcc(*'mp4v'), fps, (frame_width, frame_height))
```

```python
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Используем модель для анализа текущего кадра
    results = model(frame)
    # Фильтрация только объектов класса "person" (класс 0 в COCO)
    filtered_boxes = [box for box in results[0].boxes if int(box.cls[0]) in allowed_indices]

    # Визуализация детекций только для людей
    # for box in person_boxes:
    for i, box in enumerate(filtered_boxes):
      xyxy = box.xyxy[0]  # Координаты бокса
      conf = box.conf[0]  # Уверенность
      class_name = results[0].names[int(box.cls[0])]
      label = f'{class_name} {conf:.2f}'  # Метка с уверенностью

      # Рисование бокса и подписи на кадре
      x1, y1, x2, y2 = map(int, xyxy)
      cv2.rectangle(frame, (x1, y1), (x2, y2), (255, 0, 0), 2)
      cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)

    # Запись обработанного кадра в выходное видео
    out.write(frame)

# Завершаем обработку
cap.release()
out.release()
cv2.destroyAllWindows()
print(f"Обработанное видео сохранено в {output_path}")
```
